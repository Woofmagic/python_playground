If tensors are matrices (they aren't generally... this is a major misnomer), then we can do matrix operations.

```python
random_matrix = torch.rand(5, 5)
random_matrix
```

```python
random_matrix.T
```

```python
random_matrix.inverse()
```

```python
random_matrix.inverse () @ random_matrix
```

Interesting... It's *almost* the identity matrix. Can that error compound?

```python
random_matrix_2 = torch.rand(2, 2)
random_matrix_2
```

```python
random_matrix_2.inverse() @ random_matrix_2
```

```python
random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2
```

```python
random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2
```

```python
random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2 @ random_matrix_2.inverse() @ random_matrix_2
```

Actually, we didn't check the determinant.

```python
random_matrix_2.det()
```

```python
random_matrix_2
```

```python
random_matrix_2 = random_matrix_2 / random_matrix_2.det()
```

```python
random_matrix_2
```

```python
random_matrix_2.inverse() @ random_matrix_2
```

Welp. Yeah, low probability the determinant is not $0$ for a random square matrix.
