Python interpreter absolute path

```python
def find_python_interpreter() -> str:
    try:
        attempt_at_finding_path = sys.executable
        if (attempt_at_finding_path == ''):
            print(f"> Found empty string when finding path to Python: {attempt_at_finding_path}")
            return ''
        elif attempt_at_finding_path == None:
            print(f"> Found None when trying to find path: {attempt_at_finding_path}" )
            return ''
        else:
            print(f"> Found Python executable absolute path:\n> {attempt_at_finding_path}")
            return attempt_at_finding_path
    except:
        print(f"> Error occurred when finding Python executable path...")
        return ''

find_python_interpreter()
```
