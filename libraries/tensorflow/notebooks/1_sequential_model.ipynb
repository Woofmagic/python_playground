```python
import tensorflow as tf
import keras
from keras import layers
```

## (2): When the Use Sequential Models

```python
seq_model_1 = keras.Sequential(
    [
        layers.Dense(2, activation = 'relu', name = 'layer1'),
        layers.Dense(2, activation = 'relu', name = 'layer2'),
        layers.Dense(2, activation = 'linear', name = 'layer3'),
    ]
)

# array of Dense layers
```

```python
ones_tensor = tf.ones((3, 3))
```

```python
ones_tensor
```

```python
seq_model_1(ones_tensor)
```

How does it make sense? The input is `(3, 3)`, but the first layer has 2 nodes.

```python
ones_tensor_2 = tf.ones(
    shape = (4, 4)
)
```

```python
seq_model_1(ones_tensor_2)
```

It does care about shape somehow... I guess that `2` doesn't mean number of nodes or something...

```python
initial_model = keras.Sequential(
    [
        keras.Input(shape = (250, 250, 3)),
        layers.Conv2D(32, 5, strides = 2, activation = 'relu'),
        layers.Conv2D(32, 3, activation = 'relu'),
        layers.Conv2D(32, 3, activation = 'relu'),
    ]
)
```

```python
initial_model.inputs
```

```python
initial_model.layers
```

```python
[layer.output for layer in initial_model.layers]
```

```python
feature_extractor = keras.Model(
    inputs = initial_model.inputs,
    outputs = [layer.output for layer in initial_model.layers]
)

# this one is not so well-explained...
```

So, this model actually has several different outputs... Is that right? Recall the top of the page about inappropriate uses of Sequential models:

> A Sequential model is not appropriate when:
> - Your model has multiple inputs or multiple outputs
> ...

Maybe that's what they mean.

```python
images = tf.ones(
    shape = (1, 250, 250, 3)
)
```

```python
features = feature_extractor(images)
```
