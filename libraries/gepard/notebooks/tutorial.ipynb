```python
informed_test_datapoint = g.DataPoint(
    xB = 0.348, t = -0.3, Q2 = 3., phi = 0.3,
    process = 'ep2epgamma', # https://gepard.phy.hr/docs/observables.html
    exptype = 'fixed target',
    in1energy = 6., in1charge = -1, in1polarization = +1,
    observable = 'XS', val = 0.21, err = 0.01,
    frame = 'Trento')
```

```python
th_KM15.predict(informed_test_datapoint)
```

```python
informed_test_datapoint.W
```

```python
th_KM15.AC(informed_test_datapoint)
```

```python
th_KM15.ImH(informed_test_datapoint)
```

```python
print(th_KM15.ReH(informed_test_datapoint))
print(th_KM15.ImH(informed_test_datapoint))
print(th_KM15.ReE(informed_test_datapoint))
print(th_KM15.ImE(informed_test_datapoint))
print(th_KM15.ReHt(informed_test_datapoint))
print(th_KM15.ImHt(informed_test_datapoint))
print(th_KM15.ReEt(informed_test_datapoint))
print(th_KM15.ImEt(informed_test_datapoint))
```

The test above confirms that the KM15 GPD model sets $\text{Im}[E] = 0$ and $\text{Im}[\tilde{E}] = 0$.
