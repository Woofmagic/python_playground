```python
import iminuit
import numpy as np
from iminuit.cost import LeastSquares
from iminuit import Minuit
```

```python
print(iminuit.__version__)
```

```python
def line(x: float, α: float, β: float) -> float:
    """
    :param float x: independent variable
    :param float α: slope
    :param float β: y-intercept
    """
    return x * α + β
```

Test it

```python
# should be equal to y-intercept
line(0, 0., 5.)
```

```python
random_number = np.random.default_rng(42)
```

```python
x_data = np.linspace(-1., 1., 50)
```

```python
# should be numpy array of line-data
line(x_data, 1., 5.)
```

```python
# will be element-wise addition:
y_data_error = np.random.uniform(0., 1., 50)
```

```python
y_data = random_number.normal(
    loc = line(x_data, 1., 2.),
    scale = y_data_error
)
```

First, you need to define the objective function:

```python
least_squares = LeastSquares(
    x = x_data, 
    y = y_data,
    yerror = y_data_error,
    model = line)
```

Then, put the objective function into the Minuit fitting class *with* the initial guesses for the parameters.

```python
minuit_fitting = Minuit(least_squares, α = 1., β = 5.)
```

```python
minuit_fitting.migrad()
```

```python
minuit_fitting.hesse()
```
