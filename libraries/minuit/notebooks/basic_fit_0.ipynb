```python
import numpy as np
from iminuit import Minuit
```

```python
# The parameters of the function are the the model parameters!!!
def objective_function(a, b, c):
    return np.exp(a + b + c)
```

```python
minuit_minimizer = Minuit(objective_function, a = 10., b = -10., c = 5.)
```

```python
minuit_minimizer.fixed["a"]
```

```python
# minuit_minimizer.limits["c"] = (-10, 10)
minuit_minimizer.fixed["a"] = False
minuit_minimizer.fixed["b"] = False
minuit_minimizer.fixed["c"] = False
```

```python
minuit_minimizer.migrad()
```

```python
print(f"Parameter values: {minuit_minimizer.values}")
```

```python
print(f"Parameter errors: {minuit_minimizer.errors}")
```

```python
print(f"Function value at minimum: {minuit_minimizer.fval}")
```

```python
print(f"Covariant matrix:\n{minuit_minimizer.covariance}")
```

```python
minuit_minimizer.hesse()
```
