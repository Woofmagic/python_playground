```python
import numpy as np
import matplotlib.pyplot as plt
```

The example below is from: https://matplotlib.org/stable/gallery/mplot3d/errorbar3d.html

```python
ax = plt.figure(figsize = (10, 5.5)).add_subplot(projection='3d')


t = np.arange(0, 2*np.pi+.1, 0.01)

x, y, z = np.sin(t), np.cos(3*t), np.sin(5*t)

estep = 15

i = np.arange(t.size)

zuplims = (i % estep == 0) & (i // estep % 3 == 0)

zlolims = (i % estep == 0) & (i // estep % 3 == 2)

ax.errorbar(x, y, z, 0.2, zuplims=zuplims, zlolims=zlolims, errorevery = 1)

ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_zlabel("Z label")
```

Looks like we can't do it for a scatterplot. We need to do a hacky method:

```python
x = np.random.uniform(-1, 1, 50)

y = np.random.uniform(-1, 1, 50)

z = np.sin(x) + np.cos(y)

z_err = np.random.normal(0.1, 0.02, size=z.shape) 

figure = plt.figure(figsize=(8, 6))

axis = figure.add_subplot(111, projection='3d')

axis.scatter(x, y, z, color='blue', label='Data points')

for xi, yi, zi, zi_err in zip(x, y, z, z_err):
    axis.plot([xi, xi], [yi, yi], [zi - zi_err, zi + zi_err], color='red', alpha=0.5)

axis.set_xlabel(r'$x$')

axis.set_ylabel(r'$y$')

axis.set_zlabel(r'$z$')

axis.set_title('3D Scatter Plot with z-Error Bars')

plt.tight_layout()
plt.show()
```
